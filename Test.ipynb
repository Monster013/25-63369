```python
# Create a Python script file
script_content = f"""
import re

def find_unique_urls():
    # Specify the path to the log file
    log_file_path = '/root/cloudflared.{port}.log'

    # Read the log file
    with open(log_file_path, 'r') as file:
        log_content = file.read()

    # Define the regex pattern to find URLs
    pattern = r'https://(.*?\\.trycloudflare\\.com)'

    # Find all matching URLs and convert to a set to ensure uniqueness
    unique_urls = set(re.findall(pattern, log_content))

    # Return the formatted URLs
    return [f'Uri : https://{{url}}' for url in unique_urls]
"""

# Save the script to a Python file
with open('/content/find_urls.py', 'w') as file:
    file.write(script_content)

print("Script saved as 'find_urls.py'")
```

```python

#@title __Lyrics Convert__

import json
import os
from google.colab import files

def json_to_txt_lyrics(json_file, txt_file):
    # Open and load the JSON file
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Extract the lines
    lines = data['lyrics']['lines']

    # Open the TXT file to write
    with open(txt_file, 'w', encoding='utf-8') as f:
        for line in lines:
            # Convert startTimeMs to minutes, seconds, and milliseconds
            start_time_ms = int(line['startTimeMs'])
            minutes = start_time_ms // 60000
            seconds = (start_time_ms % 60000) // 1000
            milliseconds = start_time_ms % 1000

            # Format the time as [MM:SS.mmm]
            timestamp = f"[{minutes:02}:{seconds:02}.{milliseconds:03}]"

            # Write the timestamp and words to the TXT file
            f.write(f"{timestamp} {line['words']}\n")

# Specify the input and output files
uploaded = files.upload()
for filename in uploaded.keys():
    json_file = filename

txt_file = 'lyrics_output.txt'

# Convert the JSON to TXT
json_to_txt_lyrics(json_file, txt_file)

files.download(txt_file)
```
